```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
```

```python
X_reduced=PCA(n_components=3).fit_transform(iris.data)
Y=iris.target
Figure=plt.figure(1,figsize=(8,6))
axes=Axes3D(Figure,elev=-150,azim=110)
axes.scatter(X_reduced[:,0],X_reduced[:,1],X_reduced[:,2],c=Y,cmap=plt.cm.rainbow_r)
axes.set_title("First Three PCA Directions")
axes.set_xlabel("1st EigenVector")
axes.set_ylabel("2nd EigenVector")
axes.set_zlabel("3rd EigenVector")
```
